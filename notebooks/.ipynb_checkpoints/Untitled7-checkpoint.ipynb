```python
import pandas as pd
import numpy as np

fpathc = "/home/janekg89/Develop/Pycharm_Projects/flutype_webapp/master_2018_03_02/studies/microarray_2018_02_18/measurements/slide_22/results/quant1/circle_quality.tsv"
fpathi = "/home/janekg89/Develop/Pycharm_Projects/flutype_webapp/master_2018_03_02/studies/microarray_2018_02_18/measurements/slide_22/results/quant1/intensity.tsv"
fpaths = "/home/janekg89/Develop/Pycharm_Projects/flutype_webapp/master_2018_03_02/studies/microarray_2018_02_18/measurements/slide_22/results/quant1/std.tsv"
```

```python
this_gali = pd.read_csv(fpathi, sep='\t', index_col = 0,  dtype=str)
this_gals = pd.read_csv(fpaths, sep='\t', index_col = 0,  dtype=str)
this_galc = pd.read_csv(fpathc, sep='\t', index_col = 0,  dtype=str)
gals = [this_gali,this_gals,this_galc]
```

```python
def rows_and_cols_to_gal_file(this_gal):
    this_gal = this_gal.stack()
    id = range(1, 1 + len(this_gal.values))
    data = {"ID":id ,
            "Row": list(map(str,this_gal.index.get_level_values(0))),
            "Column": list(map(str,this_gal.index.get_level_values(1))),
            "Name":list(map(str, this_gal.values))
            }
    df = pd.DataFrame(data=data)
    df1 = df.set_index("ID")
    return df1

```

```python
this_gals = [rows_and_cols_to_gal_file(this_gal).replace([np.NaN], [None]) for this_gal in gals]
```

```python
spots =pd.DataFrame(this_gals[0]["Name"].values, columns=["intensity"])
```

```python
spots["std"] = this_gals[1]["Name"].values
spots["circle"] = this_gals[2]["Name"].values
```
