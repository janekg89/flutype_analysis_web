## Query Dataset

```python
#studies = ["microarray_2018_02_18"]
#spots = Spot.objects.filter(raw_spot__raw_spot_collection__studies__sid__in=studies)
cs = ["slide_21","slide_22","slide_23","slide_26","slide_27","slide_28","slide_31","slide_32","slide_33","slide_36","slide_37","slide_41","slide_42"]
spots = Spot.objects.filter(raw_spot__raw_spot_collection__sid__in=cs)
spots = spots.filter(spot_collection__sid = "quant1")
#spots = spots.filter(circle_quality__gte = 0.3)
```

```python
virus_batches = list(spots.values_list('raw_spot__lig_mob_batch__sid',flat=True).distinct())
```

```python
max_spots = {}
for vb in virus_batches:
    max_spots[vb] = len(spots.filter(raw_spot__lig_mob_batch__sid = vb))
    
```

```python
frames = []
cqs = np.linspace(0.00,0.90,25)
for cq in cqs:
    print(cq)
    
    this_spots = spots.filter(circle_quality__gte = cq)
    spots_pd = a.Data._reformat(this_spots)
    spots_mean = pp.mean_on_analyte_batch(spots_pd)
    frames.append(spots_mean.groupby("Analyte Batch").apply(pp.my_aggs ,max_spots))
circ_per = pd.concat(frames, keys = cqs)
circ_per = circ_per.reset_index().rename(columns={"level_0":"Circle Qual"})    
```

```python
frames = []
for cq in cqs:
    print(cq)
    this_spots = spots.filter(circle_quality__gte = cq)
    spots_pd = a.Data._reformat(this_spots)
    spots_pd = pp.outlier_filtering(spots_pd)
    spots_mean = pp.mean_on_analyte_batch(spots_pd)
    frames.append(spots_mean.groupby("Analyte Batch").apply(pp.my_aggs ,max_spots))
    
circ_per_fo = pd.concat(frames, keys = cqs)
circ_per_fo = circ_per_fo.reset_index().rename(columns={"level_0":"Circle Qual"})    
```

```python
frames = []

for cq in cqs[:15]:
    print(cq)
    this_spots = spots.filter(circle_quality__gte = cq)
    spots_pd = a.Data._reformat(this_spots)
    spots_pd = pp.outlier_filtering(spots_pd)
    spots_pd = pp.normalize(spots_pd)
    spots_mean = pp.mean_on_analyte_batch(spots_pd)
    frames.append(spots_mean.groupby("Analyte Batch").apply(pp.my_aggs ,max_spots))
circ_per_norm = pd.concat(frames, keys = cqs)
circ_per_norm = circ_per_norm.reset_index().rename(columns={"level_0":"Circle Qual"})    
```
