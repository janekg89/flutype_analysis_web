```python
frames = []
for name, data in vir1_pd.groupby(["Name"]):
    x = data.mean()
    x["intensities_mean_std"] = data["intensities"].std(ddof=1)/np.sqrt(len(data))
    x["intensities2_mean_std"] = data["intensities2"].std(ddof=1)/np.sqrt(len(data))
    
    x.name = name
    frames.append(x)
mean_data = pd.concat(frames, axis=1)
mean_data = mean_data.transpose().reset_index()
mean_data_v1 = mean_data.rename(columns={"index":"Peptide Batch"})

frames = []
for name, data in vir3_pd.groupby(["Name"]):
    x = data.mean()
    x["intensities_mean_std"] = data["intensities"].std(ddof=1)/np.sqrt(len(data))
    x["intensities2_mean_std"] = data["intensities2"].std(ddof=1)/np.sqrt(len(data))
  
    x.name = name
    frames.append(x)
mean_data = pd.concat(frames, axis=1)
mean_data = mean_data.transpose().reset_index()
mean_data_v3 = mean_data.rename(columns={"index":"Peptide Batch"})
```
